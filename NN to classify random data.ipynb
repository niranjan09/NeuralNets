```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch_size = 64
X = 2
X1 = 100
Y = 10
num_of_iterations = 5000
```

```python
#Create random data
x = np.random.randn(batch_size, X)
y = np.random.randn(batch_size, Y)
```

```python
plt.plot(x, y, 'bo')
plt.show()
```

```python
#randomly initialize weights
w11 = np.random.randn(X, X1)
w21 = np.random.randn(X1, Y)
```

```python
learning_rate = 10**(-4)
```

```python
plot_arr_loss = []
plot_arr_y = []
```

```python
for i in range(num_of_iterations):
    #forward pass
    z1 = x.dot(w11)
    a1 = np.maximum(z1, 0)
    y_pred = a1.dot(w21)

    #compute loss
    loss = np.square(y_pred - y).sum()
    plot_arr_loss.append(loss)
    print(i, loss)

    # Backprop to compute gradients of w1 and w2 with respect to loss
    gradient_y_pred = 2.0 * (y_pred - y)
    gradient_w21 = a1.T.dot(gradient_y_pred)
    gradient_a1 = gradient_y_pred.dot(w21.T)
    #gradient_z1 = grad_a1.copy()
    gradient_a1[z1 < 0] = 0
    gradient_w11 = x.T.dot(gradient_a1)

    # Update weights
    w11 -= learning_rate * gradient_w11
    w21 -= learning_rate * gradient_w21
    if i%500==0:
        learning_rate/=2
```

```python
z = x.dot(w11)
a = np.maximum(z, 0)
plot_arr_y = a.dot(w21)
```

```python
plt.plot(x, plot_arr_y, 'ro')
plt.show()
plt.plot(x, y, 'bo')
plt.show()
plt.plot(x, plot_arr_y, 'ro')
plt.plot(x, y, 'bo')
plt.show()
```

```python
plt.plot(range(num_of_iterations), plot_arr_loss)
plt.show()
```
